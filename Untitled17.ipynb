#univariate analysis


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('/content/weather-dataset.csv')
weather.head()
```

# Univariate analysis using histograms

```python
plt.figure(figsize=(16, 10))

plt.subplot(2, 3, 1)
sns.histplot(weather['Temperature (C)'], kde=True, color='skyblue')
plt.xlabel('Temperature (C)')
plt.title('Temperature Distribution')

plt.subplot(2, 3, 2)
sns.histplot(weather['Apparent Temperature (C)'], kde=True, color='salmon')
plt.xlabel('Apparent Temperature (C)')
plt.title('Apparent Temperature Distribution')

plt.subplot(2, 3, 3)
sns.histplot(weather['Humidity'], kde=True, color='green')
plt.xlabel('Humidity')
plt.title('Humidity Distribution')

plt.subplot(2, 3, 4)
sns.histplot(weather['Wind Speed (km/h)'], kde=True, color='orange')
plt.xlabel('Wind Speed (km/h)')
plt.title('Wind Speed Distribution')

plt.subplot(2, 3, 5)
sns.histplot(weather['Pressure (millibars)'], kde=True, color='purple')
plt.xlabel('Pressure (millibars)')
plt.title('Pressure Distribution')

plt.tight_layout()
plt.show()
```

# Univariate analysis using box plots

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
sns.boxplot(y=weather['Temperature (C)'], color='skyblue')
plt.ylabel('Temperature (C)')
plt.title('Temperature Box Plot')

plt.subplot(2, 3, 2)
sns.boxplot(y=weather['Apparent Temperature (C)'], color='salmon')
plt.ylabel('Apparent Temperature (C)')
plt.title('Apparent Temperature Box Plot')

plt.subplot(2, 3, 3)
sns.boxplot(y=weather['Humidity'], color='green')
plt.ylabel('Humidity')
plt.title('Humidity Box Plot')

plt.subplot(2, 3, 4)
sns.boxplot(y=weather['Wind Speed (km/h)'], color='orange')
plt.ylabel('Wind Speed (km/h)')
plt.title('Wind Speed Box Plot')

plt.subplot(2, 3, 5)
sns.boxplot(y=weather['Pressure (millibars)'], color='purple')
plt.ylabel('Pressure (millibars)')
plt.title('Pressure Box Plot')

plt.tight_layout()
plt.show()
```

# Bivariate analysis using a scatter plot

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Temperature (C)', y='Humidity', data=weather, color='skyblue')
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity')
plt.show()
```

# Bivariate analysis using a linear regression plot

```python
plt.figure(figsize=(8, 6))
sns.regplot(x='Temperature (C)', y='Humidity', data=weather, color='salmon')
plt.title('Temperature vs Humidity with Linear Regression')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity')
plt.show()
```

#Multivariate Analysis

```python
numerical_columns = ['Temperature (C)', 'Apparent Temperature (C)', 'Humidity',
                     'Wind Speed (km/h)', 'Wind Bearing (degrees)', 'Visibility (km)',
                     'Pressure (millibars)']

sns.pairplot(weather[numerical_columns])
plt.show()
```
